##Scikit-learn Design:-----

##primarily, three types of objects:----

##1.estimators:-- it estimates some parameter based on a dataset. eg. imputer. it has a fit method and transform methods. fit 
##methods- fits the dataset and calculates internal parameters.

##2.Transform:-- transform methods takes input and returns outputs based on the learning from fit(). it also conveniences ##functions called fit_transform() which fits and then transforms.

##3.predictors:-- linearregression model is an examples  of predictor .fit() and predict() are two comman function .it also ##gives score() function  which will evalute the predictions.
 

```python
## feature scaling:---
##primarily there are two type of feature scalling methods--
##1. min-max  scaling [( normalization )]--- formula--{( value - mix )/( max - min )}
 ##sklearn provides a class called  MinMaxScalar for this.
    
    
##2. [Standardization] formula---{( value - mean )/std}                     (std-standarddevition)
##sklearn provide a class called standared scalar for this.
    
    
```

```python
##Creating Pipeline..
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = "median")),## ..... add as many as you want in your pipeline.
    ('std_scaler', StandardScaler()),
    
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
```
